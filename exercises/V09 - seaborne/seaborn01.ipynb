```python
import seaborn as sns
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.set_theme(style='white')
sns.displot(tips['total_bill'], kde=True, bins=20)
#sns.histplot(tips['total_bill'])
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', kind='hex')
```

```python
sns.pairplot(tips, hue='sex', palette='coolwarm')
```

```python
sns.rugplot(tips['total_bill'])   
```

```python
sns.kdeplot(tips['total_bill'], fill=True)
sns.rugplot(tips['total_bill'], color='orange')
```

```python
sns.barplot(data=tips, x='sex', y='total_bill', hue='sex', estimator='std', palette='Spectral')
# import numpy as np
# sns.barplot(data=tips, x='sex', y='total_bill', hue='sex', estimator=np.std, palette='Spectral')
```

```python
sns.set_style('whitegrid')
sns.countplot(x='sex', data=tips, hue='sex')
```

```python
sns.boxenplot(data=tips, x='day', y='total_bill', hue='day', palette='rainbow', orient='v')
```

```python
sns.boxenplot(x='day', y='total_bill', hue='sex', data=tips, palette='coolwarm', legend=True, orient='v')
```

```python
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex', split=True)
```

```python
sns.stripplot(data=tips, x='day', y='total_bill', hue='sex')
```

```python
sns.swarmplot(data=tips, x='day', y='total_bill', hue='sex', dodge=True)
```

```python
sns.catplot(data=tips, x='day', y='total_bill', kind='bar')
```

```python
iris=sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.PairGrid(iris)
```

```python
import matplotlib.pyplot as plt
g=sns.PairGrid(iris)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
g.map_diag(sns.histplot)
```

```python
g=sns.FacetGrid(tips, col='time', row='smoker', hue='sex')
g.map(plt.scatter, 'total_bill', 'tip').add_legend()
```
